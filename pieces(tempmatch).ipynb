```python
import cv2
import numpy as np
img=cv2.imread('localised/dd5.jpg')
width = 457
height = 457
dim = (width, height)
# resize image
img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.45
thickness = 1
color = (0, 0, 169)

```

```python
#whitepawn

import numpy as np
import cv2
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
pawn_white_template = cv2.imread("temp1/wpawn.jpg",0)
w_pawn_white, h_pawn_white = pawn_white_template.shape[::-1]
res_pawn_white = cv2.matchTemplate(img_gray,pawn_white_template,cv2.TM_CCOEFF_NORMED)
threshhold = 0.795
loc = np.where(res_pawn_white >= threshhold)
ll=[]
k=[]
for pt in zip(*loc[::-1]):
    ll.append(pt)

x1=0
x2=0
for i in ll :
    x=0
    j=list(i)
    ll1=tuple(j)
    k1=abs(j[0]-x1)
    k2=abs(j[1]-x2)
   
    k=k1+k2
    if(k>8):
        cv2.rectangle(img_rgb,ll1,(ll1[0]+w_pawn_white, ll1[1]+h_pawn_white),(0,255,255),2)

        cv2.putText(img_rgb,'WP',(j[0]+35,j[1]+10), font,font_scale, color, thickness)
    x1=j[0]
    x2=j[1]
```
